###Image Addition - dodawanie obrazw


```python
img1 = cv2.imread(filename="view.jpg")
img2 = cv2.imread(filename="bear.jpg")
```

```python
add_img = cv2.add(src1=img1, src2=img2) #add azy dwa obrazy. Wane!! Obrazy musz bys tych samych wymiarw
```

```python
add_img.shape
cv2_imshow(add_img)
```

###image blending


```python
img1 = cv2.imread(filename="view.jpg")
img2 = cv2.imread(filename="bear.jpg")
```

```python
img1.shape
img2.shape
```

```python
dst = cv2.addWeighted(src1=img1, alpha=0.3, src2=img2, beta=0.7, gamma=0) 
#gamma - jasno, alpha - widoczno pierwszego obrazu, beta - widoczno drugiego obrazu 
cv2_imshow(dst)
```
