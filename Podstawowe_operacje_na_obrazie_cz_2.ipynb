```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
img = cv2.imread(filename="view.jpg")
cv2_imshow(img)
```

###wycinanie elementw

```python
left_top_square = img[:200, :200] #tu wycina si jakby tablica numpy array(numery pikseli, ktre chcemy wyci)
print(left_top_square.shape)
cv2_imshow(left_top_square)
```

```python
right_top_square = img[:200, -200:] #ten minus moemy tumaczyc jako "ile OSTATNICH pikseli"
print(right_top_square.shape)
cv2_imshow(right_top_square)
```

```python
left_bot_square = img[-200:, :200] 
print(left_bot_square.shape)
cv2_imshow(left_bot_square)
```

```python
right_bot_square = img[-200:, -200:] 
print(right_bot_square.shape)
cv2_imshow(right_bot_square)
```

###znajdowanie rodka obrazu 


```python
height, width, _ = img.shape

h0 = height // 2 - 100
h1 = height // 2 + 100
w0 = width // 2 - 100
w1 = width // 2 + 100
```

```python
middle = img[h0:h1, w0:w1]
cv2_imshow(middle)
```

###modyfikowanie obrazu 

```python
img[h0:h1, w0:w1] = 0 #zmieni wszystkie wartoci z przedziau od 0 do 200 i od 0 do 200 w np array na 0 czyli czarny
cv2_imshow(img) 
```
