```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

###progowanie

```python
img = cv2.imread(filename="bear.jpg")
gray = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY) #zmienia głębie obrazu(convert color)
cv2_imshow(gray)
```

```python
cv2_imshow(cv2.threshold(src=gray, thresh=50, maxval=255, type=cv2.THRESH_BINARY)[1])
#wszystkie wartośći w pojedynczej komórce macieży gray, które są większę od 50 zostaną zmienione na 255(biały)
```

###operacja splotu, wygładzanie

```python
kernel = np.ones(shape=(5,5), dtype=np.float32) / 25
kernel
```

```python
#to rozmywa obraz, dzięki czemu można efektywniej wykrywać krawędzie
cv2_imshow(cv2.filter2D(src=img, ddepth=-1, kernel=kernel)) 
```

```python
kernel = np.ones(shape=(5,5), dtype=np.float32) / 20
kernel
```

```python
cv2_imshow(cv2.filter2D(src=img, ddepth=-1, kernel=kernel)) 
```

```python
#to jest wykorzystywane do wykrywania krawędzi 
kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])
kernel
cv2_imshow(cv2.filter2D(src=img, ddepth=-1, kernel=kernel))# dzięki tej prostej operacji możemy  otrzymać obraz przedstawiający krawędzie zdjęcia
```

```python
kernel = np.array([[-1,0,1],
                   [-5,0,5],
                   [-1,0,1]])
kernel
cv2_imshow(cv2.filter2D(src=img, ddepth=-1, kernel=kernel))
```

###gotowe funkcje pozwalające rozmyć obraz 

```python
cv2_imshow(cv2.blur(src=img, ksize=(5,5))) #ksize - rozmiar jądra
```
