```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

###generowanie losowego obrazu


```python
#w skali szarości
array = np.random.randint(low=0, high=256, size=(400, 500)) #generuje losowa tablicę w roszmiarze 400x500 i w jednej komórce wartość od 0 do 256
array
```

```python
cv2_imshow(array)
```

```python
#kolorowy 
array = np.random.randint(low=0, high=256, size=(400, 500, 3))
array
```

```python
cv2_imshow(array)
```

```python
#czarny 
array = np.zeros(shape=(400, 500))
cv2_imshow(array)
```

```python
#biały
array = np.full(shape=(400, 500), fill_value=255)
cv2_imshow(array)
```
