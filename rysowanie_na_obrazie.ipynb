```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
bear = cv2.imread("bear.jpg")
```

###dodanie prostokąta


```python
copy = bear.copy()
cv2_imshow(cv2.rectangle(img=copy, pt1=(75, 80), pt2=(180, 200) color=(0,0,255), thickness=2)) #pt1 - lewy górny róg prostokąta, pt2 - prawy dolny róg
```

###rysowanie koła okęgu

```python
copy = bear.copy()
cv2_imshow(cv2.circle(img=copy, center=(300,300), radius=20, color=(0,0,255), thickness=-1)) #jak grubość jest na minusię to kształ się wypełnia
```

###dodanie lini

```python
copy = bear.copy()
cv2_imshow(cv2.line(img=copy, pt1=(180, 300), pt2=(300, 300), color=(0,0,255), thickness=2))
```

###dodawanie tekstu 


```python
copy = bear.copy()
cv2_imshow(cv2.putText(img=copy, text="OpenCv: Bear", org=(10, 40), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.5, color=(0,255,0), thickness=4)) #org - położenie
```
